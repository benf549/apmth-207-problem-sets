```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import json, os
```

```python
with open(os.path.join("_korali_result", "latest"), "r") as f:
    j = json.load(f)
    samples = np.array(j["Results"]["Sample Database"])
    # 90041

mean = np.mean(samples, axis=0)
mean
```

```python
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.1, zorder=10)
plt.title("5000 TMCMC Samples after 14 generations", size=24)
plt.legend(["TMCMC Samples"], fontsize=16)
plt.xlabel("a", size=16)
plt.xticks(size=16)
plt.ylabel("b", size=16)
plt.yticks(size=16)

plt.scatter(mean[0], mean[1], marker='x', color='black', s=100)
plt.gcf().set_size_inches(20, 10)
```

```python
data = pd.read_csv("data.csv")
```

```python
plt.scatter(data.x, [0 for _ in data.x], marker='x', c="black", s=100, clip_on=False, label="Samples")
plt.gcf().set_size_inches(25, 7)

plt.plot(mean[0], mean[1], 'ro', markersize=10, label="Mean Sampled Parameter Value")
plt.text(mean[0]+20, mean[1]-0.025, "Lighthouse", size=16)
plt.xlabel("a", size=16)
plt.xticks(size=16)
plt.yticks(size=16)
plt.ylabel("b", size=16)
plt.axis([-1150, 1150, 0, 1.6])
plt.vlines(mean[0], 0, mean[1], 'k', linestyle=":")
plt.plot([0, mean[0]], [0, mean[1]], ':k');
plt.legend()

plt.title(f"Mean Sampled Position for Lighthouse: \n(a, b) = ({mean[0]:0.2f}, {mean[1]:0.2f})", size=20)
```
