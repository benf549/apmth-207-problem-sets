```python
df = pd.read_csv("./histogram.csv")
x = df["x"].to_numpy()
count = df["count"].to_numpy()
plt.bar(x, count)
plt.ylabel("counts")
plt.xlabel("bin")
plt.title("Distribution of 1,000,000 points sampled from piecewise CDF (100 bins)")
plt.gcf().set_size_inches(20, 10)
plt.savefig("histogram.jpg")
```
