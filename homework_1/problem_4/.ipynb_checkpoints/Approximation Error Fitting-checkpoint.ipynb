```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
teqs = ["mc", "is", "rs"]
ds = [1, 2, 4, 8, 16]
Ms = [1, 10, 100, 1_000, 10_000, 100_000, 1_000_000]
colors = ["r", "g", "b", "c", "m", "y", "gray"]

analytical = [d/3 for d in ds]
```

# Plot Results of Programs

```python
teq = "mc"
idx = 1
fig, ((a1, a2, a3, a4), (a5, a6, a7, a8)) = plt.subplots(2, 4)
fig.set_size_inches(20, 10)
axs = [a1, a2, a3, a4, a5, a6, a7]
for m, col, ax in zip(Ms, colors, axs):
    xs = []
    ys = []
    yerrs = []
    for d, true in zip(ds, analytical):
        file = f"{teq}_{d}_{m}_100_trials.csv"
        data = np.array(pd.read_csv(file, header=None)[0])
        error = np.std(data)
        mean = np.mean(data)
        xs.append(d)
        ys.append(mean)
        yerrs.append(error)
        
    ax.scatter(xs, ys, color=col, )
    ax.errorbar(xs, ys, linestyle='None', color=col, yerr=yerrs, capsize=6.0)
    ax.scatter(xs, analytical, color='k', marker="x")
    ax.set_xlabel("d")
    ax.set_ylabel("Approximate value of Integral")
    ax.legend(["Monte Carlo Estimate", "Analytical"])
    idx += 1
    ax.set_title("M=" + str(m))
plt.tight_layout()
plt.savefig("monte_carlo_results.jpg")
```

# Create Error Fit Plots

```python
from scipy.optimize import curve_fit

idx = 1

def error_fit(x, b, c):
    return b*1/np.sqrt(x) + c

fit_x = np.logspace(0, 6, 100)
teq = "mc"
dcols = ["r", "g", "b", "c", "m"]
fig, ((a1, a2), (a3, a4), (a5, a6)) = plt.subplots(3, 2)
fig.set_size_inches(20, 10)
axs = [a1, a2, a3, a4, a5]
print(axs)

for d, dcol, true, ax in zip(ds, dcols, analytical, axs):
    xs = []
    ys = []
    for m, col in zip(Ms, colors):
        file = f"{teq}_{d}_{m}_100_trials.csv"
        data = np.array(pd.read_csv(file, header=None)[0])
        error = np.std(data)
        mean = np.mean(data)
        
        # plt.plot(fit_x, error_fit(fit_x))
        xs.append(m)
        ys.append(error)
        
    p, pcov = curve_fit(error_fit, xs, ys)
    pc_err = np.sqrt(np.diag(pcov))
    print(p, pc_err)
    ax.scatter(xs, ys, color=dcol)
    ax.plot(fit_x, error_fit(fit_x, *p), "k:")
    ax.set_xscale('log')
    ax.set_xlabel("M")
    ax.set_ylabel("error")
    ax.set_title(f"Error vs M for d={d}")
    ax.legend(['Measured Error', f'NLLS Fit to a*1/sqrt(M) + b\na={p[0]:.2} ± {pc_err[0]:.1}\nb={p[1]:0.2} ± {pc_err[1]:.1}'])
    idx += 1

plt.suptitle("Monte Carlo Integration Results (100 trials)", fontsize=16)
plt.tight_layout()

plt.savefig("monte_carlo_error_analysis.jpg")
```

# Timing Python Equivalent

```python
def function_to_integrate(x, d):
    output = 0.0
    output = np.sum(x.T @ x)
    # for i in x:
    #     output += (i**2)
    return (2 ** (-1 * d)) * output

def g_function(sv, interval_max, interval_min, d):
    if np.any(interval_min > sv) or np.any(sv > interval_max):
        return 0.0
    
    return function_to_integrate(sv, d)

def q_function(sv, interval_max, interval_min, d):
    C = 1.0 
    if np.any(interval_min > sv) or np.any(sv > interval_max):
        return 0.0

    # The vector fell within the volume so return 1.
    return C * (0.1 + function_to_integrate(sv, d))

def generate_gaussian_sample_vector(d):
    return np.random.normal(0, 0.5, size=(d))

def rejection_sampling_integration(d, M, interval_max, interval_min):
    
    edge_length = interval_max - interval_min
    V = edge_length ** d

    num_sum = 0.0
    den_sum = 0.0
    accepted = 0

    while accepted < M:
        sv = generate_gaussian_sample_vector(d)
        q = q_function(sv, interval_max, interval_min, d)
        g = g_function(sv, interval_max, interval_min, d)
        if g < 0.0001:
            continue
        else:
            # if accepted % 100 == 0:
            #     print(accepted, accepted / M)
            accepted += 1
            w_i = 1.0 / q
            num_sum += g * w_i
            den_sum += w_i
    
    return V * num_sum / den_sum
```

```python
%%time
for m in Ms:
    print(rejection_sampling_integration(1, m, 1, -1))
```
