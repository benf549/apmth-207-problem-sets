```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from scipy.optimize import curve_fit
```

```python
M = 100
N = 1000

def random_walk(M, N, d):
    data = np.zeros((M, d, N+1))
    for n in range(1, N+1):
        data[:, :, n] = data[:, :, n-1] # Copy previous state.
        idx = np.random.randint(d, size=(M, 1)) # Choose which dimension index to move in
        inc = np.random.choice([-1, 1], size=(M, 1)) # Choose whether to go forward or backwards

        for i, j in enumerate(idx):
            data[i,j,n] += inc[i]
    return data

d2 = random_walk(M, N, 2)
d3 = random_walk(M, N, 3)
```

```python
ns = [x for x in range(1, N+1)]

def distance_from_origin(data, n):
    d_bar = np.mean(np.linalg.norm(data[:, :, n], axis=1))
    return d_bar

def dbar_fxn(n, a, b):
    return b*(n**a)

d_bar_2 = [distance_from_origin(d2, n) for n in ns]
d_bar_3 = [distance_from_origin(d3, n) for n in ns]

plt.plot(ns, d_bar_2, 'b', label=r"$\overline{d}$ (2D)")
plt.xlim(0, N)
popt, pcov = curve_fit(dbar_fxn, ns, d_bar_2)
perr = np.sqrt(np.diag(pcov))
xs = np.arange(0, N, 0.5)
ys = dbar_fxn(xs, *popt)
plt.plot(xs, ys, 'k', label=r"2D NLLS Fit" + "\n" + f"a={popt[0]:0.3}±{perr[0]:0.1}, b={popt[1]:0.3}±{perr[1]:0.1}")

plt.plot(ns, d_bar_3, 'r', label=r"$\overline{d}$ (3D)")
popt, pcov = curve_fit(dbar_fxn, ns, d_bar_3)
perr = np.sqrt(np.diag(pcov))
xs = np.arange(0, N, 0.5)
ys = dbar_fxn(xs, *popt)
plt.plot(xs, ys, 'k-.', label=r"3D NLLS Fit" + "\n" + f"a={popt[0]:0.3}±{perr[0]:0.1}, b={popt[1]:0.3}±{perr[1]:0.1}")

plt.gcf().set_size_inches(20, 10)
plt.xlabel("number of steps in random walk (n)", fontsize=20)
plt.ylabel(r"average distance of random walker from origin ( $\overline{d}$ )", fontsize=20)
plt.legend(fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title("Least Squares Fitting of 2D vs 3D Random Walk Distance From Origin\n" + r"M=100, N=1000", fontsize=25)
plt.tight_layout()
plt.savefig("2b.jpg")
```

```python
rs = [x for x in np.arange(11)]
def find_n_for_R(data, R):
    taus = np.zeros(M)
    if R == 0:
        return taus
    for n in range(1, N+1):
        ds = np.linalg.norm(data[:, :, n], axis=1)
        for ddx, d in enumerate(ds):
            if d >= R:
                if taus[ddx] == 0:
                    taus[ddx] = n
    return taus

def compute_avg_num_steps(taus, R):
    if R != 0:
        filtered = [x for x in taus if x != 0]
        return np.mean(filtered), np.std(filtered)
    else:
        return np.mean(taus), np.std(taus)

def compute_tau_bars(data, rs):
    tau_bars = []
    tau_stds = []
    for r in rs:
        taus = find_n_for_R(data, r)
        tau_bar, tau_std = compute_avg_num_steps(taus, r)
        tau_bars.append(tau_bar)
        tau_stds.append(tau_std)
    return tau_bars, tau_stds

tau_bars_2d, tau_stds_2d = compute_tau_bars(d2, rs)
tau_bars_3d, tau_stds_3d = compute_tau_bars(d3, rs)

# Plot 2D Case
plt.plot(rs, tau_bars_2d, 'bo', label=r"$\overline{\tau}$ (2D)")
plt.errorbar(rs, tau_bars_2d, linestyle='None', color='b', yerr=tau_stds_2d, capsize=10.0, label="2D Std. Dev.")
popt, pcov = curve_fit(dbar_fxn, rs, tau_bars_2d)
perr = np.sqrt(np.diag(pcov))
xs = np.arange(0, 12, 0.5)
ys = dbar_fxn(xs, *popt)
plt.plot(xs, ys, 'b--', alpha=0.5, label=r"2D NLLS Fit" + "\n" + f"a={popt[0]:0.3}±{perr[0]:0.1}, b={popt[1]:0.3}±{perr[1]:0.1}")

# Plot 3D Case
plt.plot(rs, tau_bars_3d, 'ro', label=r"$\overline{\tau}$ (3D)")
plt.errorbar(rs, tau_bars_3d, linestyle='None', color='r', yerr=tau_stds_3d, capsize=10.0, label="3D Std. Dev.")
popt, pcov = curve_fit(dbar_fxn, rs, tau_bars_3d)
perr = np.sqrt(np.diag(pcov))
xs = np.arange(0, 12, 0.5)
ys = dbar_fxn(xs, *popt)
plt.plot(xs, ys, 'r--', alpha=0.5, label=r"3D NLLS Fit" + "\n" + f"a={popt[0]:0.3}±{perr[0]:0.1}, b={popt[1]:0.3}±{perr[1]:0.1}")

plt.legend(fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(0, 11)
plt.xlabel("Radius From Origin (R)", fontsize=20)
plt.ylabel(r"First Passage Time ( $\overline{\tau}$ )", fontsize=20)
plt.title("Least Squares Fitting of 2D vs 3D Random Walker First Passage Time \n as a Function of Origin-Centered Sphere Radius (M=100, N=1000)", fontsize=25)

plt.gcf().set_size_inches(20, 10)
plt.tight_layout()
plt.savefig("2c.jpg")
```
