```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import scipy.stats as stat
from scipy.optimize import curve_fit
```

```python
# Initialize PDF Functions from which to compute alpha
def gaussian_pdf(x, mu, C):
    return stat.multivariate_normal.pdf(x, mean=mu, cov=C)

def cauchy_pdf(x, loc, gamma):
    return stat.cauchy.pdf(x, loc=loc, scale=gamma)

def sample_cauchy(prev, g):
    return stat.cauchy.rvs(loc=prev, scale=g)

def propose_next_state(prev_vec, gamma):
    return np.array([sample_cauchy(x, gamma) for x in prev_vec])


def run_metropolis_hastings(mu, Sigma, gamma):
    
    N_iter = 1000
    xs_array = np.zeros((N_iter, 2))
    num_rejected = 0

    x_n = np.array([0, 0])
    for idx in range(N_iter):
        xs_array[idx, :] = x_n
        # Compute proposed transition
        x_prime = propose_next_state(x_n, gamma)

        # Compute alpha
        numerator = gaussian_pdf(x_prime, mu, Sigma) * cauchy_pdf(x_n[0], loc=x_prime[0], gamma=gamma) * cauchy_pdf(x_n[1], loc=x_prime[1], gamma=gamma)
        denominator = gaussian_pdf(x_n, mu, Sigma) * cauchy_pdf(x_prime[0], loc=x_n[0], gamma=gamma) * cauchy_pdf(x_prime[1], loc=x_n[1], gamma=gamma) 
        alpha = min(1, numerator / denominator)

        # Draw random number to determine if alpha accepted.
        rand = np.random.uniform(0, 1)
        if rand < alpha:
            # Transition Accepted
            x_n = x_prime 
        else:
            # Transition Rejected
            num_rejected += 1
    return xs_array, num_rejected


mu = [0,0]

gamma_no_cov = 0.5
p1 = 0
sigma1 = [[1, p1], [p1, 1]]
MH_no_covariance, no_cov_rej = run_metropolis_hastings(mu, sigma1, gamma_no_cov)

gammas1 = np.arange(0.1, 2, 0.1)
no_cov_rejections = [run_metropolis_hastings(mu, sigma1, x)[1] for x in gammas1]

gamma_cov = 0.085
p2 = 99/100
sigma2 = [[1, p2], [p2, 1]]
MH_big_covariance, no_cov_rej = run_metropolis_hastings(mu, sigma2, gamma_cov)

gammas2 = np.arange(0.01, 0.2, 0.01)
big_cov_rejections = [run_metropolis_hastings(mu, sigma2, x)[1] for x in gammas2]
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.set_title(f"Metrop. Hastings Sampling of Multivariate Gaussian from \n Independent Cauchy Distributions, gamma={gamma_no_cov}, N=1000, p=0", fontsize=18)
t = np.linspace(-4, 4, 100)
h = np.linspace(-4, 4, 100)
w_0, w_1 = np.meshgrid(t, h)
z = stat.multivariate_normal.pdf(np.dstack((w_0, w_1)), mean=mu, cov=sigma1)
ax1.contour(t, h, z, cmap="Greys", linewidths=3, alpha=1)
ax1.plot(MH_no_covariance[:, 0], MH_no_covariance[:, 1], 'o', color='forestgreen', alpha=0.3, label="MH Sampled Point")
ax1.set_xlabel("X", fontsize=15)
ax1.set_ylabel("Y", fontsize=15)
ax1.legend(fontsize=15, title="True Multivariate Gaussian Shown as Contour Plot")

ax2.set_title(f"Metrop. Hastings Sampling of Multivariate Gaussian from \n Independent Cauchy Distributions, gamma={gamma_cov}, N=1000, p=99/100", fontsize=18)
t = np.linspace(-2, 2, 100)
h = np.linspace(-2, 2, 100)
w_0, w_1 = np.meshgrid(t, h)
z = stat.multivariate_normal.pdf(np.dstack((w_0, w_1)), mean=mu, cov=sigma2)
ax2.contour(t, h, z, cmap="Greys", linewidths=3, alpha=1)
ax2.plot(MH_big_covariance[:, 0], MH_big_covariance[:, 1], 'o', color='forestgreen', alpha=0.3, label="MH Sampled Point")
ax2.set_xlabel("X", fontsize=15)
ax2.set_ylabel("Y", fontsize=15)
ax2.legend(fontsize=15, title="True Multivariate Gaussian Shown as Contour Plot")

ax3.plot(gammas1, np.array(no_cov_rejections)/1000, "o-", label="Measured Rejection Data")
ax3.set_title(f"Rejection Rate as Function of Gamma for p = {p1}", fontsize=18)
ax3.set_xlabel("Gamma Value", fontsize=15)
ax3.set_ylabel("Number Rejected / Number Sampled", fontsize=15)
ax3.hlines(0.5, 0, 5, color="k", linestyles='dashed', label="50% Rejection Threshold")
ax3.vlines(gamma_no_cov, 0, 1, color="r", linestyles='solid', label="gamma used above")
ax3.set_xlim(0.1, 1.9)
ax3.set_ylim(0, 1)
ax3.legend(fontsize=15)

ax4.plot(gammas2, np.array(big_cov_rejections)/1000, "o-", label="Measured Rejection Data")
ax4.set_title(f"Rejection Rate as Function of Gamma for p = {p2}", fontsize=18)
ax4.set_xlabel("Gamma Value", fontsize=15)
ax4.set_ylabel("Number Rejected / Number Sampled", fontsize=15)
ax4.hlines(0.5, 0, 0.15, color="k", linestyles='dashed', label="50% Rejection Threshold")
ax4.vlines(gamma_cov, 0, 1, color="r", linestyles='solid', label="gamma used above")
ax4.set_xlim(0.01, 0.15)
ax4.set_ylim(0, 1)
ax4.legend(fontsize=15)


fig.set_size_inches(20, 10)
plt.tight_layout()
plt.savefig("Q3_metropolis_hastings.jpg")
```
